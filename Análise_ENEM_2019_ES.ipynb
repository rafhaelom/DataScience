# Análise do ENEM 2019 - 2º Edição Imersão Dados Alura

```
# Importando Bibliotecas.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Lendo dataset.
dados = pd.read_csv("https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true")
dados.head()
```

```
# Quantidade de linhas e colunas.
dados.shape
```

# FILTRANDO OS DADOS DO ESTADO DO ESPÍRITO SANTO (ES).

```
# Selecionando apenas os dados do ES.
dados_es = dados[dados.SG_UF_RESIDENCIA == "ES"]
dados_es.head()
```

```
# Quantidade de linhas e colunas.
dados_es.shape
```

```
# Nome das colunas.
dados_es.columns.values
```

```
# Quantidade de inscritos no ES.
len(dados_es["NU_INSCRICAO"])
```

```
# Idades dos inscritos no ES.
dados_es["NU_IDADE"].value_counts().sort_index()
```

```
# Histograma da distribuição de idades dos inscritos.
dados_es["NU_IDADE"].hist(bins = 20, figsize = (10,8));
```

```
# Quantidade de inscritos por sexo.
dados_es["TP_SEXO"].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(16,8))
```

```
# Quantidade de inscritos por estado civil.
dados_es["TP_ESTADO_CIVIL"].value_counts().sort_index().plot(kind='bar', figsize=(8,6))
```

```
# Quantidade de inscritos por cor e raça.
dados_es["TP_COR_RACA"].value_counts().sort_index().plot(kind='bar', figsize=(10,8))
```

```
# Quantidade de inscritos por nacionalidade.
dados_es["TP_NACIONALIDADE"].value_counts().sort_index().plot(kind='bar', figsize=(8,6))
```

```
dados_es["NO_MUNICIPIO_RESIDENCIA"].value_counts().sort_index().plot(kind='bar', figsize=(12,10))
```

```
renda_ordenada = dados_es["Q006"].unique()
renda_ordenada.sort()
```

```
dados_es["NU_NOTA_TOTAL"] = dados_es[provas].sum(axis=1)
dados_es.head()
```

```
provas = ["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]
dados_es[provas].describe()
```

```
dados_es[provas].boxplot(grid=True, figsize= (10,8))
plt.title("Distribuição das notas no ES");
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(x="Q006", y = "NU_NOTA_TOTAL", data = dados_es, order = renda_ordenada)
plt.title("Boxplot das notas de total pela renda")
```

```
dados_nao_treineiros = dados_es.query('IN_TREINEIRO == 0')['Q006'].value_counts(normalize=True)
dados_sim_treineiros = dados_es.query('IN_TREINEIRO == 1')['Q006'].value_counts(normalize=True)

df = pd.DataFrame({'Treineiros': dados_nao_treineiros,
                   'Não Treineiros': dados_sim_treineiros})

df.plot.bar(figsize=(10,8))
plt.title("Treineiros e não treineiros pela renda");
```

```
dados_sem_notas_zero = dados_es.query("NU_NOTA_TOTAL != 0")
dados_sem_notas_zero.head()
```

```
correlacao = dados_sem_notas_zero[provas].corr()
correlacao
```

```
sns.heatmap(correlacao, cmap="Blues", center=0, annot=True)
```
