Desafio 3: Criar uma função para plotar o boxplot do seaborn.

```
dados.head()
```

```
# Função para plotar boxplot do seaborn com renda ordenada por notas.
def plotarBoxplot(x_data, y_data, fonte_data, title):
  plt.figure(figsize=(10, 6))
  sns.boxplot(x= x_data, y = y_data, data = fonte_data, order = renda_ordenada)
  plt.title(title)
```

```
# Gráfico da renda por NU_NOTA_TOTAL.
plotarBoxplot(x_data = "Q006", y_data = "NU_NOTA_TOTAL", fonte_data = dados_sem_notas_zero, title = "Boxplot das notas de total pela renda")
```

```
# Gráfico da renda por NU_NOTA_MT.
plotarBoxplot(x_data = "Q006", y_data = "NU_NOTA_MT", fonte_data = dados_sem_notas_zero, title = "Boxplot das notas NU_NOTA_MT pela renda")
```

```
# Gráfico da renda por NU_NOTA_LC.
plotarBoxplot(x_data = "Q006", y_data = "NU_NOTA_LC", fonte_data = dados_sem_notas_zero, title = "Boxplot das notas NU_NOTA_LC pela renda")
```

```
# Gráfico da renda por NU_NOTA_CN.
plotarBoxplot(x_data = "Q006", y_data = "NU_NOTA_CN", fonte_data = dados_sem_notas_zero, title = "Boxplot das notas NU_NOTA_CN pela renda")
```

```
# Gráfico da renda por NU_NOTA_CH.
plotarBoxplot(x_data = "Q006", y_data = "NU_NOTA_CH", fonte_data = dados_sem_notas_zero, title = "Boxplot das notas NU_NOTA_CH pela renda")
```
